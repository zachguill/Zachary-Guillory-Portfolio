```python
# From the Seaborn library load mpg dataset

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = sns.load_dataset('mpg')
%matplotlib inline
```

```python
#1) First find the number of unique values in cylinders column and then plot a piechart of this column.
headings =[]
values =[]
def createHeadings():
    cy = df['cylinders'].unique()
    print(cy)
    for i in cy:
        headings.append(i)
        values.append((df['cylinders'] == i).sum())

createHeadings()
plt.pie(values, labels = headings)
plt.title('Distribution of Cylinders')
plt.show()
```

```python
# 2) Make subplots of scatter plots depicting the relationship between horsepower and mpg and another one between weight and mpg.
HP = df['horsepower']
MPG= df['mpg']
WGT = df['weight']

"""
print(df['horsepower'].max())
print(df['horsepower'].min())
print(df['mpg'].max())
print(df['mpg'].min())
print(df['weight'].max())
print(df['weight'].min())
"""
plt.scatter(HP, MPG)
plt.xlim(40,240)
plt.ylim(5,50)
plt.title('Horsepower vs Miles Per Gallon')
plt.xlabel('Horse Power')
plt.ylabel('MPG')
plt.show()

plt.scatter(WGT, MPG)
plt.xlim(1500,5200)
plt.ylim(5,50)
plt.title('Weight vs Miles Per Gallon')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```

```python
# 3) Plot histogram to show horsepower distribution.
plt.hist(HP)
plt.title('Distribution of Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Number of Occurances')
plt.show()
```
